```python
def preprocess_data(data, endpoint):
    processed_data = []
    if "issues" in endpoint:
        for item in data:
            processed_item = {
                'number': item.get('number'),
                'id': item.get('id'),
                'title': item.get('title'),
                'state': item.get('state'),
                'created_at': item.get('created_at'),
                'closed_at': item.get('closed_at'),
                'author': preprocess_data(item.get('user', {}), 'author'),
                'author_association': item.get('author_association'),
                'assignee': preprocess_data(item.get('assignee', {}), 'author'),
                'labels': preprocess_data(item.get('labels', []), 'labels'),
                #'comments': item.get('comments')
            }
            processed_data.append(processed_item)
    elif endpoint == 'commits':

        for item in data:
            processed_item = {
                'commit': {
                    'author': {
                        'name': item.get('commit', {}).get('author', {}).get('name'),
                        'email': item.get('commit', {}).get('author', {}).get('email'),
                        'date': item.get('commit', {}).get('author', {}).get('date')
                    },
                    'message': item.get('commit', {}).get('message')
                },
                'author': preprocess_data(item.get('author', {}), 'author'),
                'committer': preprocess_data(item.get('committer', {}), 'author'),
                'parents': [parent.get('sha') for parent in item.get('parents', [])]
            }
            processed_data.append(processed_item)
    elif endpoint == 'branches':
        for item in data:
            processed_item = {
                'name': item.get('name')
            }
            processed_data.append(processed_item)
    elif endpoint == 'pulls':
        for item in data:
            processed_item = {
                'id': item.get('id'),
                'number': item.get('number'),
                'state': item.get('state'),
                'title': item.get('title'),
                'created_at': item.get('created_at'),
                'updated_at': item.get('updated_at'),
                'closed_at': item.get('closed_at'),
                'merged_at': item.get('merged_at'),
                'author': preprocess_data(item.get('user', { }), 'author'),
                'assignee': preprocess_data(item.get('assignee', {}), 'author'),
                'requested_reviewers': [preprocess_data(reviewer.get('login'), 'author') for reviewer in item.get('requested_reviewers', [])],
                'labels': preprocess_data(item.get('labels', []), 'labels'),
                'draft': item.get('draft'),
                'merged': item.get('merged'),
                'mergeable': item.get('mergeable'),
                'comments': item.get('comments'),
                'review_comments': item.get('review_comments')
            }
            processed_data.append(processed_item)
    elif endpoint == 'languages':
        # Languages endpoint returns a dict of language:bytes pairs
        # Convert to list of dicts for consistency
        for language, bytes_count in data.items():
            processed_item = {
                'language': language,
                'bytes': bytes_count
            }
            processed_data.append(processed_item)
    elif endpoint == 'subscribers':
        for item in data:
            processed_item = {
                'login': item.get('login'),
                'id': item.get('id'),
                'type': item.get('type'),
                'site_admin': item.get('site_admin')
            }
            processed_data.append(processed_item)
    elif endpoint == 'author':
        if isinstance(data, dict):

            return {
                'login': data.get('login'),
                'id': data.get('id'),
                'type': data.get('type'), 
                'user_view_type': data.get('type'),
                'site_admin': data.get('site_admin')
            }
        # Return None for any other case
        return None
    elif endpoint == 'labels':
        for item in data:
            processed_item = {
                'name': item.get('name'),
                'description': item.get('description')
            }
            processed_data.append(processed_item)
    elif endpoint == 'commits':
        for item in data:
            processed_item = {
                'sha': item.get('sha'),
                'author': item.get('commit', {}).get('author', {}).get('name'),
                'date': item.get('commit', {}).get('author', {}).get('date'),
                'message': item.get('commit', {}).get('message')
            }
            processed_data.append(processed_item)
    elif endpoint == 'pull_requests':
        for item in data:
            processed_item = {
                'id': item.get('id'),
                'number': item.get('number'),
                'title': item.get('title'),
                'body': item.get('body'),
                'state': item.get('state'),
                'created_at': item.get('created_at'),
                'updated_at': item.get('updated_at'),
                'closed_at': item.get('closed_at'),
                'merged_at': item.get('merged_at'),
                'author': preprocess_data(item.get('user', {}), 'author'),
                'assignee': preprocess_data(item.get('assignee', {}), 'author'),
                'requested_reviewer': preprocess_data(item.get('requested_reviewers', [{}])[0].get('login'), 'author'),
                'additions': item.get('additions'),
                'deletions': item.get('deletions'),
                'changed_files': item.get('changed_files')
            }
            processed_data.append(processed_item)
    elif endpoint == 'contributors':
        for item in data:
            processed_item = {
                'login': item.get('login'),
                'id': item.get('id'),
                'contributions': item.get('contributions'),
                'type': item.get('type'),
                'user_view_type': item.get('type'),
                'site_admin': item.get('site_admin')
            }
            processed_data.append(processed_item)
    elif endpoint == 'forks':
        # Process fork data including basic repository info and owner details
        for item in data:
            processed_item = {
                'id': item.get('id'),
                'full_name': item.get('full_name'),
                'description': item.get('description'),
                'owner': preprocess_data(item.get('owner', {}), 'author'),
                'created_at': item.get('created_at'),
                'updated_at': item.get('updated_at'),
                'pushed_at': item.get('pushed_at'),
                'stargazers_count': item.get('stargazers_count'),
                'watchers_count': item.get('watchers_count'),
                'forks_count': item.get('forks_count'),
                'default_branch': item.get('default_branch')
            }
            processed_data.append(processed_item)
    elif endpoint == 'releases':
        for item in data:
            processed_item = {
                'id': item.get('id'),
                'tag_name': item.get('tag_name'),
                'name': item.get('name'),
                'body': item.get('body'),
                'draft': item.get('draft'),
                'prerelease': item.get('prerelease'),
                'created_at': item.get('created_at'),
                'published_at': item.get('published_at'),
                'author': preprocess_data(item.get('author', {}), 'author')
            }
            processed_data.append(processed_item)
    elif endpoint == 'issues/comments':
        for item in data:
            processed_item = {
                'id': item.get('id'),
                'body': item.get('body'),
                'body_text': item.get('body_text'), 
                'body_html': item.get('body_html'),
                'created_at': item.get('created_at'),
                'updated_at': item.get('updated_at'),
                'user': preprocess_data(item.get('user', {}), 'author'),
                'author_association': item.get('author_association'),
                'reactions': {
                    'total_count': item.get('reactions', {}).get('total_count', 0),
                    '+1': item.get('reactions', {}).get('+1', 0),
                    '-1': item.get('reactions', {}).get('-1', 0), 
                    'laugh': item.get('reactions', {}).get('laugh', 0),
                    'hooray': item.get('reactions', {}).get('hooray', 0),
                    'confused': item.get('reactions', {}).get('confused', 0),
                    'heart': item.get('reactions', {}).get('heart', 0),
                    'rocket': item.get('reactions', {}).get('rocket', 0),
                    'eyes': item.get('reactions', {}).get('eyes', 0)
                }
            }
            processed_data.append(processed_item)
    elif endpoint == 'pulls/comments':
        for item in data:
            processed_item = {
                'id': item.get('id'),
                'body': item.get('body'),
                'created_at': item.get('created_at'),
                'updated_at': item.get('updated_at'),
                'user': preprocess_data(item.get('user', {}), 'author'),
                'author_association': item.get('author_association'),
                'reactions': {
                    'total_count': item.get('reactions', {}).get('total_count', 0),
                    '+1': item.get('reactions', {}).get('+1', 0),
                    '-1': item.get('reactions', {}).get('-1', 0),
                    'laugh': item.get('reactions', {}).get('laugh', 0), 
                    'hooray': item.get('reactions', {}).get('hooray', 0),
                    'confused': item.get('reactions', {}).get('confused', 0),
                    'heart': item.get('reactions', {}).get('heart', 0),
                    'rocket': item.get('reactions', {}).get('rocket', 0),
                    'eyes': item.get('reactions', {}).get('eyes', 0)
                }
            }
            processed_data.append(processed_item)
    
    if processed_data:
        return processed_data
    else:
        return data
```

```python
def update_repository_data(collection, repo_name, endpoint, page_data, page_count):
    if '?' in endpoint:
        endpoint = "closed_issues"   
    
    processed_data = preprocess_data(page_data, endpoint)

    if page_count == 0:
        # For first page, initialize array
        collection.update_one(
            {'name': repo_name},
            {'$set': {endpoint: processed_data}},
            upsert=True
        )
    else:
        # For subsequent pages, append to array
        collection.update_one(
            {'name': repo_name},
            {'$push': {endpoint: {'$each': processed_data}}}
        )
```

```python
# Function to handle pagination and data collection for any endpoint
def fetch_and_insert_paginated_data(repo_name, endpoint, headers, repository_collection):
    url = f'https://api.github.com/repos/{repo_name}/{endpoint}'
    # Add per_page parameter
    url = f'{url}{"&" if "?" in url else "?"}per_page=100'
    has_next_page = True
    page_count = 0

    while has_next_page:
        response = requests.get(url, headers=headers)

        if response.status_code == 200:
            # Get data from current page
            page_data = response.json()
            
            if page_data:  # Only process if we have data
                try:
                    update_repository_data(repository_collection, repo_name, endpoint, page_data, page_count)
                    print(f"Added page {page_count + 1} with {len(page_data)} items for {endpoint}")
                    page_count += 1
                except Exception as e:
                    print(f"Error updating page {page_count + 1} of {endpoint} for {repo_name}: {str(e)}")
                    break



            # Check for next page in Link header
            if 'Link' in response.headers:
                links = response.headers['Link'].split(', ')
                next_link = None
                for link in links:
                    if 'rel="next"' in link:
                        next_link = link[link.find('<')+1:link.find('>')]
                        break
                
                if next_link:
                    url = next_link
                else:
                    has_next_page = False
            else:
                has_next_page = False

        else:
            print(f"Error fetching {endpoint} for {repo_name}: {response.status_code}")
            break

endpoints = {
    'branches': 'branches',
    'commits': 'commits',
    'issues': 'issues',
    'pulls': 'pulls', 
    'contributors': 'contributors',
    'languages': 'languages',
    'closed_issues': 'issues?state=closed',
    'forks': 'forks',
    'repo_info': '', 
    'subscribers': 'subscribers',
    'releases': 'releases',
    'issue_comments': 'issues/comments',
    'pull_comments': 'pulls/comments',
    'collaborators': 'collaborators'
}

# Headers for GitHub API
headers = {
    'Accept': 'application/vnd.github+json',
    'Authorization': f'token',
    'X-GitHub-Api-Version': '2022-11-28'
}
db_names = ["repos_java", "repos_c", "repos_cpp"]
for name in db_names:
    db = client[name]
# Create repository collection
    repository_collection = db['repositories']

    if db.name == "repos_java":
        repositories = repositories_java
    elif db.name == "repos_c":
        repositories = repositories_c
    elif db.name == "repos_cpp":
        repositories = repositories_cpp

    # Loop through each repository
    for repo_name in repositories:
        print(f"Processing {repo_name}...")

        # Loop through each endpoint
        for data_type, endpoint in endpoints.items():
            print(f"Fetching {data_type}...")

            # Fetch and insert data page by page
            fetch_and_insert_paginated_data(repo_name, endpoint, headers, repository_collection)

        print(f"Completed processing {repo_name}")
```
